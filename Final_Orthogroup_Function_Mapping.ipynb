### Getting the Functions for All OrthoGroups from Closely related species (Using OrthoFinder output)

```python
## Reading and renaming the columns 
import pandas as pd 
OrthoGroup = pd.read_excel("Orthogroups-fixFormat-2.xlsx")
OrthoGroupDF = OrthoGroup.rename(columns={"Unnamed: 0":"Orthogroup"})
OrthoGroupDF
```

## Using OrthoGroupDF and UniportOutput DataFrame to get Missing UniportIDs that does not have any function in UniportOutput.

```python
def getMissingOrthogroup(primary, secondary):
    OrthoGroup_GeneName = primary
    UniportOutput = secondary
    UniportOutput_MissingFrom_OrthoGroup_GeneName = []

    OrthoGroup_GeneName_Orthogroup = list(OrthoGroup_GeneName['Orthogroup'])
    UniportOutput_Orthogroup = list(UniportOutput['Orthogroup'])

    for i in range(len(OrthoGroup_GeneName_Orthogroup)): # Len of bigger df to avoid index out of bound expection
        if OrthoGroup_GeneName_Orthogroup[i] not in UniportOutput_Orthogroup:
            UniportOutput_MissingFrom_OrthoGroup_GeneName.append(OrthoGroup_GeneName_Orthogroup[i])

    UniportOutput_MsnFrm_OrthoGroup_GeneName = pd.DataFrame(UniportOutput_MissingFrom_OrthoGroup_GeneName, columns = ['Orthogroup'])
    return UniportOutput_MsnFrm_OrthoGroup_GeneName
```

```python
OrthoGroup_Missing_Uniport = getMissingOrthogroup(OrthoGroupDF, UniportOutput)
OrthoGroup_Missing_Uniport
```

```python
# Now we are taking out the records of Missing_uniport1 from the main Orhthogroup table
Missing_uniport3 = pd.merge(OrthoGroup_Missing_Uniport, OrthoGroupDF, how="inner", on="Orthogroup")
Missing_uniport3
```

## Using each of the output file from OrthoFinder (Five Species) and getting the GeneNames from OrthoGroupDF and from that we are getting the related functions from each species output files(ex. Aedes_aegypti.xlsx, that have gene name and functions)

```python
Aedes_aegypti = pd.read_excel("Aedes_aegypti.xlsx")
Aedes_aegypti = Aedes_aegypti.drop("Unnamed: 0", 1)
Aedes_aegypti
Aedes_aegypti_dict = {}
for i in range(len(Aedes_aegypti)):
    Aedes_aegypti_dict[Aedes_aegypti[0][i]] = Aedes_aegypti[1][i]

Aedes_aegypti
```

```python
def createOrthoFunctionDict(masterDFname, newDict, masterDFColName):
    resultDict = {}
    for i in range(len(masterDFname[masterDFColName])):
        if not isinstance(masterDFname[masterDFColName][i],float):
            for element in masterDFname[masterDFColName][i].split(','):
                if element in newDict:
                    resultDict[masterDFname["Orthogroup"][i]] = [newDict[element.strip()]]
                    resultDict[masterDFname["Orthogroup"][i]].append(newDict[element.strip()])
    
    func = resultDict.values()
    NewDF = pd.DataFrame({"Orthogroup": resultDict.keys(),"Function [CC]":func})

    OrthoGroup_GeneName_IntMissing = getMissingOrthogroup(masterDFname, NewDF)
    OrthoGroup_GeneName_Int = pd.merge(masterDFname, NewDF, on= 'Orthogroup')
    
    return (OrthoGroup_GeneName_IntMissing, OrthoGroup_GeneName_Int)
```

```python
## To get all the common orthogroups with functions 
OrthoGroup_GeneName_IntMiss, OrthoGroup_GeneName_Int = createOrthoFunctionDict(Missing_uniport3, Aedes_aegypti_dict, "Aedes_aegypti") 
OrthoGroup_GeneName_Int = OrthoGroup_GeneName_Int.rename(columns={"Function [CC]":"Function"})
OrthoGroup_GeneName_Int
```

```python
## to get the remaining OrthoGroup IDs without any functions from Aedes_aegypti output
## using that output for next species so we will avoid repeatation
OrthoGroup_GeneName_PresentAA = getMissingOrthogroup(Missing_uniport3, OrthoGroup_GeneName_Int)
OrthoGroup_GeneName_AA = pd.merge(Missing_uniport3, OrthoGroup_GeneName_PresentAA, on= 'Orthogroup')
OrthoGroup_GeneName_AA
```

```python
Anopheles_gambiae = pd.read_excel("Anopheles_gambiae.xlsx")
Anopheles_gambiae = Anopheles_gambiae.drop("Unnamed: 0", 1)
Anopheles_gambiae
Anopheles_gambiae_dict = {}
for i in range(len(Anopheles_gambiae)):
    Anopheles_gambiae_dict[Anopheles_gambiae[0][i]] = Anopheles_gambiae[1][i]
```

```python
OrthoGroup_GeneName_IntAGMiss, OrthoGroup_GeneName_IntAG = createOrthoFunctionDict(OrthoGroup_GeneName_AA, Anopheles_gambiae_dict, "Anopheles_gambiae") 
OrthoGroup_GeneName_IntAG = OrthoGroup_GeneName_IntAG.rename(columns={"Function [CC]":"Function"})
OrthoGroup_GeneName_IntAG
```

```python
OrthoGroup_GeneName_PresentAG = getMissingOrthogroup(OrthoGroup_GeneName_AA, OrthoGroup_GeneName_IntAG)
OrthoGroup_GeneName_AG = pd.merge(OrthoGroup_GeneName_AA, OrthoGroup_GeneName_PresentAG, on= 'Orthogroup')
OrthoGroup_GeneName_AG
```

```python
Culex_quinquefasciatus = pd.read_excel("Culex_quinquefasciatus.xlsx")
#Culex_quinquefasciatus
Culex_quinquefasciatus = Culex_quinquefasciatus.drop("Unnamed: 0", 1)
Culex_quinquefasciatus
Culex_quinquefasciatus_dict = {}

for i in range(len(Culex_quinquefasciatus)):
    Culex_quinquefasciatus_dict[Culex_quinquefasciatus[0][i]] = Culex_quinquefasciatus[1][i]
    
Culex_quinquefasciatus_dict
```

```python
OrthoGroup_GeneName_IntCQMiss, OrthoGroup_GeneName_IntCQ = createOrthoFunctionDict(OrthoGroup_GeneName_AG, Culex_quinquefasciatus_dict, "Culex_quinquefasciatus") 
OrthoGroup_GeneName_IntCQ = OrthoGroup_GeneName_IntCQ.rename(columns={"Function [CC]":"Function"})
OrthoGroup_GeneName_IntCQ
```

```python
OrthoGroup_GeneName_PresentCQ = getMissingOrthogroup(OrthoGroup_GeneName_AG, OrthoGroup_GeneName_IntCQ)
OrthoGroup_GeneName_CQ = pd.merge(OrthoGroup_GeneName_AG, OrthoGroup_GeneName_PresentCQ, on= 'Orthogroup')
OrthoGroup_GeneName_CQ
```

```python
Drosophila_melanogaster = pd.read_excel("Drosophila_melanogaster.xlsx")
Drosophila_melanogaster = Drosophila_melanogaster.drop("Unnamed: 0", 1)
Drosophila_melanogaster
Drosophila_melanogaster_dict = {}

for i in range(len(Drosophila_melanogaster)):
    Drosophila_melanogaster_dict[Drosophila_melanogaster[0][i]] = Drosophila_melanogaster[1][i]
```

```python
OrthoGroup_GeneName_IntDMMiss, OrthoGroup_GeneName_IntDM = createOrthoFunctionDict(OrthoGroup_GeneName_CQ, Drosophila_melanogaster_dict, "Drosophila_melanogaster") 
OrthoGroup_GeneName_IntDM = OrthoGroup_GeneName_IntDM.rename(columns={"Function [CC]":"Function"})
OrthoGroup_GeneName_IntDM
```

```python
OrthoGroup_GeneName_PresentDM = getMissingOrthogroup(OrthoGroup_GeneName_CQ, OrthoGroup_GeneName_IntDM)
OrthoGroup_GeneName_DM = pd.merge(OrthoGroup_GeneName_CQ, OrthoGroup_GeneName_PresentDM, on= 'Orthogroup')
OrthoGroup_GeneName_DM
```

```python
Wyeomyia_samithii = pd.read_excel("Wyeomyia_samithii.xlsx")
Wyeomyia_samithii = Wyeomyia_samithii.drop("Unnamed: 0", 1)
Wyeomyia_samithii
Wyeomyia_samithii_dict = {}

for i in range(len(Wyeomyia_samithii)):
    Wyeomyia_samithii_dict[Wyeomyia_samithii[0][i]] = Wyeomyia_samithii[1][i]
```

```python
OrthoGroup_GeneName_IntWSMiss, OrthoGroup_GeneName_IntWS = createOrthoFunctionDict(OrthoGroup_GeneName_DM, Wyeomyia_samithii_dict, "Wyeomyia_smithii") 
OrthoGroup_GeneName_IntWS = OrthoGroup_GeneName_IntWS.rename(columns={"Function [CC]":"Function"})
OrthoGroup_GeneName_IntWS
```
