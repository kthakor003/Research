```python
### reading Excel file as dataframe

import pandas as pd
OrthoFinder = pd.read_excel("Input_unique.IDs.xlsx")
```

```python
Orthogroup = []
Drosophila_melanogaster = []
for rowNum in range(len(OrthoFinder)):
    if len(OrthoFinder["Drosophila_melanogaster"][rowNum]) > 1:
        result = OrthoFinder["Drosophila_melanogaster"][rowNum].split(',')
        for flybaseId in result:
            Orthogroup.append(OrthoFinder["Orthogroup"][rowNum])
            Drosophila_melanogaster.append(flybaseId)
    else:
        Orthogroup.append(OrthoFinder["Orthogroup"][rowNum])
        Drosophila_melanogaster.append(OrthoFinder["Drosophila_melanogaster"][rowNum])
        
#print("Orthogroup", Orthogroup)
#print("Drosophila_melanogaster", Drosophila_melanogaster)
```

```python
## Dropping the duplicates and getting the final 8319 orthogroups IDs 
OrthoFinder = pd.DataFrame()
OrthoFinder["Orthogroup"] = Orthogroup
OrthoFinder["Drosophila_melanogaster"] = Drosophila_melanogaster
OrthoFinder["Orthogroup"].drop_duplicates()
```

```python
## reading the Flybase output file with Annotation symbol(Common between Flybase output and Uniport poutput)
FlybaseIDs = pd.read_excel("FlyBase_CG_IDs.xlsx")
FlybaseIDs["ANNOTATION_SYMBOL"] = FlybaseIDs["ANNOTATION_SYMBOL"].str.split('-').str[0]
FlybaseIDs
```

```python
### inner join of Orthogroup IDs and Flybase IDs on common column with FlybaseIDsabs
GeneGroup_FlybaseIDs = pd.merge(OrthoFinder, FlybaseIDs,  how ='inner', left_on=['Drosophila_melanogaster'], right_on = ['#SUBMITTED ID'])
GeneGroup_FlybaseIDs.drop_duplicates().reset_index(drop=True) ## Dropping duplicates
GeneGroup_FlybaseIDs
```

```python
## Reading the Uniport Output file so we can use respective Functions from here for each OrthoGroups 
Uniport_output = pd.read_excel("uniprot-yourlist_newest.tab.xls")
Uniport_output = Uniport_output.rename(columns = {'yourlist:M202109104ABAA9BC7178C81CEBC9459510EDDEA31C8D66A': 'Gene IDs'})
Uniport_output
```

```python
geneIdProtNameFunc = {}
print(Uniport_output["Gene IDs"][1])
uniPort = pd.DataFrame()
for i in range(len(Uniport_output)):
    if Uniport_output["Gene IDs"][i] in geneIdProtNameFunc:
        if Uniport_output["Protein names"][i] not in geneIdProtNameFunc[Uniport_output["Gene IDs"][i]]:
            geneIdProtNameFunc[Uniport_output["Gene IDs"][i]].append(Uniport_output["Protein names"][i])
            geneIdProtNameFunc[Uniport_output["Gene IDs"][i]].append(Uniport_output["Function [CC]"][i])
    else:
        geneIdProtNameFunc[Uniport_output["Gene IDs"][i]] = [Uniport_output["Protein names"][i]]
        geneIdProtNameFunc[Uniport_output["Gene IDs"][i]].append(Uniport_output["Function [CC]"][i])
```

```python
flybaseUniport_df = pd.merge(GeneGroup_FlybaseIDs, Uniport_output,  how='inner', left_on=['ANNOTATION_SYMBOL'], right_on = ['Gene IDs'])
flybaseUniport_df = flybaseUniport_df.drop_duplicates().reset_index(drop=True)
```

```python
flybaseUniportdf = flybaseUniport_df[["Orthogroup","#SUBMITTED ID","Gene IDs","Protein names","Function [CC]"]]
flybaseUniportdf
```

```python
geneIdOrthoProtNameFunc = {}
# print(flybaseUniportdf["#SUBMITTED ID"][1])

for i in range(len(flybaseUniportdf)):
#     print(flybaseUniport_df["Orthogroup"][i])
    if flybaseUniportdf["Orthogroup"][i] in geneIdOrthoProtNameFunc:
        if not isinstance(flybaseUniportdf["Function [CC]"][i], float):
            geneIdOrthoProtNameFunc[flybaseUniportdf["Orthogroup"][i]][0].append(flybaseUniportdf["#SUBMITTED ID"][i])
            geneIdOrthoProtNameFunc[flybaseUniportdf["Orthogroup"][i]][1].append(flybaseUniportdf["Gene IDs"][i])
            geneIdOrthoProtNameFunc[flybaseUniportdf["Orthogroup"][i]][2].append(flybaseUniportdf["Protein names"][i])
            geneIdOrthoProtNameFunc[flybaseUniportdf["Orthogroup"][i]][3].append(flybaseUniportdf["Function [CC]"][i])
    elif not isinstance(flybaseUniportdf["Function [CC]"][i], float):
        geneIdOrthoProtNameFunc[flybaseUniportdf["Orthogroup"][i]] = [[flybaseUniportdf["#SUBMITTED ID"][i]]]
        geneIdOrthoProtNameFunc[flybaseUniportdf["Orthogroup"][i]].append([flybaseUniportdf["Gene IDs"][i]])
        geneIdOrthoProtNameFunc[flybaseUniportdf["Orthogroup"][i]].append([flybaseUniportdf["Protein names"][i]])
        geneIdOrthoProtNameFunc[flybaseUniportdf["Orthogroup"][i]].append([flybaseUniportdf["Function [CC]"][i]])

flybase = []
geneId = []
protnames = []
func = []
for i in geneIdOrthoProtNameFunc.values():
    flybase.append(i[0])
    geneId.append(i[1])
    protnames.append(i[2])
    func.append(i[3])

flybase[0]
```

####  put orthogroups as main column and join all other column relevant to it. so we will have one ortho group,  gene_IDs, one FB_ID, multiple protein name or num and one or more function or may not. (geneIdOrthoProtNameFunc) put all this column as a dataframe

```python
dataframe = pd.DataFrame(columns=["Orthogroup", "Flybase ID", "Gene ID", "Protein names", "Function [CC]"]) 

pd.set_option('display.max_columns', None)
df = pd.DataFrame({"Orthogroup": geneIdOrthoProtNameFunc.keys(),
                   'Flybase ID': flybase,
                   "Gene ID" :geneId,
                   "Protein names": protnames,
                   "Function [CC]":func
                  })
df
```
